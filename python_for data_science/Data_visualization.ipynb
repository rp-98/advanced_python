```python
import numpy as np
from numpy import random
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig=plt.figure()
```

```python
ax1 = fig.add_subplot(2, 2, 1)
```

```python
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(random.randn(50).cumsum(), 'k--') 
```

```python
plt.hist(random.randn(100), bins=20, color='k', alpha=0.3)
```

```python
plt.scatter(np.arange(30), np.arange(30) + 3 *random. randn(30)) 
```

```python
fig = plt.figure() 
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='r', alpha=0.3) 
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3) 
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g', alpha=0.5)
ax.add_patch(rect) 
ax.add_patch(circ) 
ax.add_patch(pgon)
```

**Plotting in pandas**

```python
from pandas import DataFrame, Series
import pandas as pd
s = Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = DataFrame(np.random.randn(10, 4).cumsum(0),columns=['A', 'B', 'C', 'D'],index=np.arange(0, 100, 10))
df.plot()
```

```python
df = DataFrame(np.random.rand(6, 4),index=['one', 'two', 'three', 'four', 'five', 'six'],columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df
```

```python
df.plot(kind='bar')
```
